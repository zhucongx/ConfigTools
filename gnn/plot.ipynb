```python
# %matplotlib inline
import pandas as pd
pd.options.mode.chained_assignment = None  # default='warn'
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "Times New Roman"
import matplotlib
from sklearn.metrics import mean_squared_error
```

```python
test_pred = np.genfromtxt('test.pred', delimiter=',')
train_pred = np.genfromtxt('train.pred', delimiter=',')
test_true = np.genfromtxt('test.true', delimiter=',')
train_true = np.genfromtxt('train.true', delimiter=',')
```

```python
mean_squared_error(test_true,test_pred)
```

```python
mean_squared_error(train_true,train_pred)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
lim_range = [0, 1.2]

plt.plot(np.linspace(lim_range[0], lim_range[1], num=1000), np.linspace(lim_range[0], lim_range[1], num=1000), color='grey', ls='--',lw=5)

plt.scatter(train_true, train_pred, color="orange",alpha=.75, label='training data')
plt.scatter(test_true, test_pred, color="blue",alpha=.75, label='validation data')

plt.xlim(lim_range)
plt.ylim(lim_range)
ax.set_aspect('equal', adjustable='box')
for tick in ax.xaxis.get_major_ticks():
    tick.label.set_fontsize(25) 
for tick in ax.yaxis.get_major_ticks():
    tick.label.set_fontsize(25)
ax.set_ylabel('predicted migration barrier (eV)', fontdict={'family' : 'Times New Roman', 'size':25})

ax.legend(prop={'size':22})

plt.savefig('model.png', dpi=300, bbox_inches="tight",transparent=True)
```
